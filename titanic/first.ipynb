```python
import pandas as pd
import numpy as np
```

### Load train and test data.

```python
train_df = pd.read_csv('inputs/train.csv')
test_df = pd.read_csv('inputs/test.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_df.describe()
```

### # of lack of value

```python
print(train_df.isnull().sum())
```

### Fill 'Age' with mean and 'Embarked' with mode, delete 'Cabin' column.

```python
age_mean = train_df.mean()['Age']
train_df['Age'] = train_df['Age'].fillna(age_mean)
test_df['Age'] = test_df['Age'].fillna(age_mean)
```

```python
filled = train_df.drop('Cabin', axis = 1)
filled_test = test_df.drop('Cabin', axis = 1)
```

```python
mode_embarked =  filled['Embarked'].mode()[0]
filled['Embarked'].fillna(mode_embarked, inplace = True)
filled_test['Embarked'].fillna(mode_embarked, inplace = True)
```

```python
filled.isnull().sum()
```

```python
filled_test.isnull().sum()
```

```python
fare_mean = train_df.mean()['Fare']
filled_test['Fare'] = filled_test['Fare'].fillna(fare_mean)
```

```python
filled = filled.drop(['PassengerId', 'Name', 'Ticket'], axis = 1)
filled_test = filled_test.drop(['PassengerId', 'Name', 'Ticket'], axis = 1)
```

### Convert categorical classes into one-hot vectors

```python
filled.head()
```

```python
sex_onehot = pd.get_dummies(filled['Sex'])
embarked_onehot = pd.get_dummies(filled['Embarked'])
onehotted = pd.concat((filled, sex_onehot, embarked_onehot), axis = 1)

sex_onehot_test = pd.get_dummies(filled_test['Sex'])
embarked_onehot_test = pd.get_dummies(filled_test['Embarked'])
onehotted_test = pd.concat((filled_test, sex_onehot_test, embarked_onehot_test), axis = 1)
```

```python
onehotted = onehotted.drop('Sex', axis = 1)
onehotted = onehotted.drop('female', axis = 1)

onehotted_test = onehotted_test.drop('Sex', axis = 1)
onehotted_test = onehotted_test.drop('female', axis = 1)
```

```python
onehotted = onehotted.drop('Embarked', axis = 1)
onehotted = onehotted.drop('Q', axis = 1)

onehotted_test = onehotted_test.drop('Embarked', axis = 1)
onehotted_test = onehotted_test.drop('Q', axis = 1)
```

```python
onehotted.head()
```

```python
X = onehotted.values[:, 1:]
y = onehotted.values[:, 0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
print('traininig set :', classifier.score(X_train, y_train))
print('test set :', classifier.score(X_test, y_test))
```

```python
prediction = classifier.predict(onehotted_test.values)
```

```python
submission = pd.DataFrame({
        "PassengerId": test_df["PassengerId"],
        "Survived": prediction
    })
```

```python
submission.to_csv('submission.csv', index=False)
```
